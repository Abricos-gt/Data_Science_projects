# 20. Classify Movies Based on Ratings [Excellent, Good, and Average]


```python
df.columns
```

```python
def rating(Rating):
    if Rating >= 7.0:
        return "Excellent"
    elif Rating >= 6.0:
        return "Good"
    else :
        return "Average"
```

```python
df['Rating_cat'] = df['Rating'].apply(rating)
```
